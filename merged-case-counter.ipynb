```python
import pandas as pd

# Assuming you have already calculated the maximum values
max_values = [
    phenonix_df['total_cases'].max(),
    chicago_df['total_cases'].max(),
    san_antonio_df['total_cases'].max(),
    ny_df['total_cases'].max(),
    la_df['total_cases'].max(),
    houston_df['total_cases'].max()
]

# List of city names
cities = ['Phoenix', 'Chicago', 'San Antonio', 'New York', 'Los Angeles', 'Houston']

# Create a DataFrame with 'city' and 'max_total_cases' columns
max_values_df = pd.DataFrame({'city': cities, 'max_total_cases': max_values})

# Read the income data from the CSV file
income_df = pd.read_csv('/Users/ben_nicholson/Desktop/household_income.csv')

# Read the population data from the new CSV file
population_df = pd.read_csv('/Users/ben_nicholson/Visual Code Projects/Adv Database Final Project/daily_case_counter/most_populated_cities - usa_top_10_most_populated_cities.csv')

# Merge max_values_df with income_df based on the 'city' column
result_df = pd.merge(max_values_df, income_df[['city', 'individual_median_income', 'household_median_income']], on='city')

# Merge the result with population_df based on the 'city' column
result_df = pd.merge(result_df, population_df[['City', 'Pop2020', 'density']], left_on='city', right_on='City')

# Drop the duplicated 'City' column
result_df = result_df.drop('City', axis=1)

# Assuming your DataFrame is named result_df
result_df['max_total_cases'] = result_df['max_total_cases'].astype(int)

# Format the printed tables
pd.set_option('display.float_format', '{:,.2f}'.format)  # Format float numbers with commas
pd.set_option('display.max_columns', None)  # Display all columns
pd.set_option('display.width', 1000)  # Set a wider display width

# Print the updated DataFrame with the 'max_total_cases' column converted to int
print(result_df.to_string(index=False))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set(style="whitegrid")

# Create a bar plot for max total cases
plt.figure(figsize=(10, 6))
sns.barplot(x='city', y='max_total_cases', data=result_df, palette='viridis')
plt.title('Max Total Cases by City')
plt.xlabel('City')
plt.ylabel('Max Total Cases')
plt.show()

# Create a bar plot for individual median income
plt.figure(figsize=(10, 6))
sns.barplot(x='city', y='individual_median_income', data=result_df, palette='magma')
plt.title('Individual Median Income by City')
plt.xlabel('City')
plt.ylabel('Individual Median Income')
plt.show()

# Create a bar plot for household median income
plt.figure(figsize=(10, 6))
sns.barplot(x='city', y='household_median_income', data=result_df, palette='plasma')
plt.title('Household Median Income by City')
plt.xlabel('City')
plt.ylabel('Household Median Income')
plt.show()

# Create a scatter plot for population density vs. max total cases
plt.figure(figsize=(10, 6))
sns.scatterplot(x='density', y='max_total_cases', data=result_df, hue='city', palette='Set2', s=100)
plt.title('Population Density vs. Max Total Cases')
plt.xlabel('Population Density')
plt.ylabel('Max Total Cases')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have the ny_df, la_df, phenonix_df, chicago_df, and houston_df DataFrames
# (Replace these with the actual variable names you have used)

# Combine the data into a single DataFrame
combined_df = pd.concat([ny_df, la_df, phenonix_df, chicago_df, houston_df], keys=['NYC', 'LA', 'Phoenix', 'Chicago', 'Houston'])

# Convert 'date' column to datetime format
combined_df['date'] = pd.to_datetime(combined_df['date'], format='%m/%d/%Y', errors='coerce')

# Sort DataFrame by 'date'
combined_df = combined_df.sort_values(by='date')

# Set the time period
start_date = pd.Timestamp('2020-03-01')
end_date = pd.Timestamp('2020-08-31')  # Updated to August 31, 2020

# Filter data for the specified time period
combined_df = combined_df.loc[(combined_df['date'] >= start_date) & (combined_df['date'] <= end_date)]

# Plotting
plt.figure(figsize=(16, 8))

# Plot the time series for each city
for city, city_df in combined_df.groupby(level=0):
    plt.plot(city_df['date'], city_df['new_case'], label=f'{city} - New Cases')

# Set the maximum y-value dynamically
plt.ylim(0, combined_df['new_case'].max() + 100)

# Vertical line representing when the universal mask mandate was implemented in NYC
nyc_mask_mandate_date = pd.to_datetime('04/17/2020', format='%m/%d/%Y', errors='coerce')
plt.axvline(x=nyc_mask_mandate_date, color='r', linestyle='--', label='NYC Mask Mandate')

# Vertical lines representing when mask mandates were implemented in other cities
la_mask_mandate_date = pd.to_datetime('07/10/2020', format='%m/%d/%Y', errors='coerce')
plt.axvline(x=la_mask_mandate_date, color='g', linestyle='--', label='LA Mask Mandate')

phoenix_mask_mandate_date = pd.to_datetime('03/30/2020', format='%m/%d/%Y', errors='coerce')
plt.axvline(x=phoenix_mask_mandate_date, color='b', linestyle='--', label='Phoenix Mask Mandate')

chicago_mask_mandate_date = pd.to_datetime('05/01/2020', format='%m/%d/%Y', errors='coerce')
plt.axvline(x=chicago_mask_mandate_date, color='m', linestyle='--', label='Chicago Mask Mandate')

houston_mask_mandate_date = pd.to_datetime('07/03/2020', format='%m/%d/%Y', errors='coerce')
plt.axvline(x=houston_mask_mandate_date, color='c', linestyle='--', label='Houston Mask Mandate')

plt.title('Comparison of COVID-19 Cases in Different Cities')
plt.xlabel('Date')
plt.ylabel('New Cases')
plt.legend()
plt.grid(True)
plt.show()
```
